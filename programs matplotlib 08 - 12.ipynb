```python
import matplotlib.pyplot as plt
#initializing the data
x = [10,20,30,40]
y = [20,25,35,55]
fig = plt.figure(figsize = (5,4))
ax = fig.add_axes([1,1,1,1])
ax1 = ax.plot(x,y)
ax2 = ax.plot(y,x)
ax.set_title("Linear graph")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend(labels = ('line 1','line 2'))
plt.show()
```

```python

import matplotlib.pyplot as plt
#initializing the data
x = [10,20,30,40]
y = [20,25,35,55]
#creating a new figure with width = 5 inches
#and height = 4 inches
fig = plt.figure(figsize=(5,4))
#creating first axes for the figure
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
#creating second axes for the figure
ax2 = fig.add_axes([1,0.1,0.8,0.8])
#adding the data to be plotted
ax1.plot(x,y)
ax2.plot(y,x)
plt.show()
```

```python
import matplotlib.pyplot as plt
#initializing the data
x = [10,20,30,40]
y = [20,25,35,55]
axes1=plt.subplot2grid((7,1),(0,0),rowspan = 2,colspan = 1)
axes2 = plt.subplot2grid((7,1),(2,0),rowspan = 2,colspan = 1)
#plotting the data
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#histogram
import matplotlib.pyplot as plt
#data to display on plots
x = [1,2,3,4,5,6,7,2]
#this will plot a simple histogram
plt.hist(x, bins = [1,2,3,4,5,6,7])
#title to the plot
plt.title("Histogram",color="red")
#adding the elements
plt.legend(["Hist"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2,2.5,3,3.5,4.5,4.7,5.0]
y = [7.5,8,8.5,9,9.5,10,10.5]
x1 = [9,8.5,9,9.5,10,10.5,12]
y1 = [3,3.5,4.7,4,4.5,5,5.2]
plt.scatter(x, y, label='high income low saving',color='g')
plt.scatter(x1, y1, label='low income high savings',color='r')
plt.xlabel('savings*100')
plt.ylabel('income*1000')
plt.title('SCATTER PLOT')
plt.legend()
plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt
#data to display on plots
x = [1,2,3,4]
#this will explode the wedge
#i.e will seperate the 1st wedge
#from the chart
e = (0.2,0,0,0)
#this will plot a single pie chart
plt.pie(x, explode = e)
#title to the plot
plt.title("PIE CHART")
plt.show()
```

```python
#3d plot
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]
#creating the figure objects
fig = plt.figure()
#keeping the projection = 3d
#ctreats the 3d plot
ax = plt.axes(projection = '3d')
ax.plot3D(z,y,x)
```
